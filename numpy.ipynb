# Execution time differences

## Lists 

```python
import time
```

```python
revenues  = [5000, 10000, 15000, 20000, 25000]
```

```python
initial_time1 = time.time()
sum = 0

for i in revenues:
  sum += i

print(sum)
termination_time1 = time.time()

execution_time1 = termination_time1 - initial_time1
print("Execution time: ", execution_time1)
```

## NumPy

```python
array2 = np.array(revenues)
array2

initial_time2 = time.time()
sum = array2.sum()

print(sum)
termination_time2 = time.time()

execution_time2 = termination_time2 - initial_time2

print("Execution time: ", execution_time2)
```

```python
difference = execution_time1 / execution_time2
difference
```

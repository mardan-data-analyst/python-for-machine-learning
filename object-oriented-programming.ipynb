# Manually changing varibles

```python
class EmployeeClass1():
  employee_name = "Ben"
  department = "Sales"
  starting_year = 2020
  salary = 5000
```

```python
myObject1 = EmployeeClass1()
```

```python
myObject1.department
```

```python
print(myObject1.employee_name)
print(myObject1.department)
print(myObject1.starting_year)
print(myObject1.salary)
```

# __init__ method

```python
class EmployeeClass2():
  def __init__(self, employee_name, department, starting_year, salary):
    self.employee_name = employee_name
    self.department = department
    self.starting_year = starting_year
    self.salary = salary
```

```python
employee_1 = EmployeeClass2("Ben", "Sales", 2020, 5000)
```

```python
print(employee_1.employee_name)
print(employee_1.department)
print(employee_1.starting_year)
print(employee_1.salary)
```
